## Consolidate the files 

```
import csv
import os
import glob
import pandas as pd

'''
To run this script, you must save the files you want ot consolidate in a subfolder called 'input'.
The 'input' subfolder must be located in the same directory as this script.
Once the script is complete, the consolidated data will be saved as 'pandasOutput.tsv' in the 'output' folder.
'''

#/*------define file extension ------*/
extension = 'tsv'

#/*------define the location of input folder and the file name and location of the output ------*/
#cwd = os.getcwd()
#os.chdir(cwd+'/input')
cwd = '/content'
#os.chdir('/content')
outputFile = cwd+'/output/'+'pandasOutput'+'.'+extension

#/*-----look for all files with the file extension (defined above).------*/
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

#/*-----combine all files in the list------*/
data = pd.concat([pd.read_csv(f) for f in all_filenames ])

#/*-----export to csv------*/
data.to_csv(outputFile, index=False, encoding='utf-8-sig')


```
